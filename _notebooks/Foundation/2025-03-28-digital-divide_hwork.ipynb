```python
def explore_digital_divide():
    # Ask the user for their choice of metric
    metric = input(
        "Which digital divide metric would you like to explore?\n" +
        "1: Income-based internet access\n" +
        "2: Urban vs rural access\n" +
        "3: Age-based internet usage\n" +
        "4: Global connectivity rates\n"
    )

    # Sample data
    income_data = {
        "categories": ["Lowest 20%", "Second 20%", "Middle 20%", "Fourth 20%", "Highest 20%"],
        "values": [62, 71, 80, 88, 95]
    }

    location_data = {
        "categories": ["Urban", "Rural"],
        "values": [94, 83]
    }

    age_data = {
        "categories": ["18-29", "30-49", "50-64", "65+"],
        "values": [97, 93, 88, 61]
    }

    global_data = {
        "categories": ["North America", "Europe", "Asia Pacific", "Latin America", "Middle East/Africa"],
        "values": [90, 87, 54, 68, 40]
    }

    # Select dataset based on user choice
    data = None
    title = None

    if metric == "1":
        data = income_data
        title = "Internet Access by Income Level (%)"
    elif metric == "2":
        data = location_data
        title = "Urban vs Rural Internet Access (%)"
    elif metric == "3":
        data = age_data
        title = "Internet Usage by Age Group (%)"
    elif metric == "4":
        data = global_data
        title = "Internet Access by Region (%)"
    else:
        print("Invalid selection!")
        return

    # Display the data
    print(title)
    print("=" * len(title))

    max_value = max(data["values"])
    max_bar_length = 40

    for i in range(len(data["categories"])):
        bar_length = round((data["values"][i] / max_value) * max_bar_length)
        bar = "‚ñà" * bar_length
        print(f"{data['categories'][i].ljust(12)}: {bar} {data['values'][i]}%")

# Run the function to explore different aspects of the digital divide
explore_digital_divide()
```
